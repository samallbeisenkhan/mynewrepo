**1.1.** RNN — реттілігі бар деректерді өңдейтін желі. Оның басты ерекшелігі — алдыңғы қадам ақпараты hidden state арқылы келесі қадамға беріледі, сондықтан модель контексті сақтай алады. Feedforward желіде мұндай ішкі жады жоқ, әр кіріс бөлек өңделеді.

**1.2.** RNN мәтін классификациясында, машиналық аудармада, сөйлеуді тануда, уақыттық қатарды болжауда, келесі сөзді не символды болжауда тиімді қолданылады. Себебі мұндай есептерде элементтердің реті мен алдыңғы контекст маңызды.

**1.3.** Hidden state — алдыңғы уақыт қадамдарынан жиналған ақпараттың ішкі көрінісі. Әр қадамда жаңа кіріс (x_t) пен алдыңғы (h_{t-1}) бірігіп, жаңа (h_t) есептеледі, сондықтан модель қысқа мерзімді контексті есте ұстайды.
(h_t = \tanh(W_x x_t + W_h h_{t-1} + b))

**1.4.** Vanishing gradient кезінде градиент уақыт бойымен кері тарағанда өте кішірейіп кетеді де, ертерек қадамдар іс жүзінде үйренбей қалады. Exploding gradient кезінде керісінше градиент шамадан тыс өсіп, оқыту тұрақсызданады. Бұл мәселе RNN-де бірдей салмақтардың уақыт бойынша көп рет қолданылуынан күшейеді.

**1.5.** LSTM мен GRU ұзақ тәуелділіктерді жақсырақ сақтау үшін ұсынылды. LSTM арнайы қақпалар мен cell state арқылы ұзақ мерзімді ақпаратты сақтайды; оның негізгі идеясы — қателік ағынын ұзақ уақыт бойы өшірмей жеткізу. GRU де gating механизмін қолданады, бірақ құрылымы жеңілірек: update және reset қақпалары арқылы маңызды ақпаратты өткізіп, артық ақпаратты тежейді.

**1.6.** RNN-ді өте ұзын тізбектерде, жоғары деңгейде параллель есептеу қажет болғанда, сондай-ақ Transformer жақсы нәтиже беретін қазіргі NLP есептерінде қолдану тиімсіз болуы мүмкін. Себебі RNN қадам-қадаммен есептейді, сондықтан оқу уақыты ұзарады және ұзақ контексті ұстау қиындайды.

**2.3.** Ұзын тізбектерде градиент нөлге жақындайды, себебі BPTT кезінде градиент көптеген туындылардың көбейтіндісі түрінде шығады. Егер әр қадамдағы көбейткіш 1-ден кіші болса, олардың ұзын тізбектегі көбейтіндісі өте кішірейеді. Sigmoid үшін
(\sigma'(x) = \sigma(x)(1 - \sigma(x))),
оның ең үлкен мәні 0.25. Tanh үшін
(\tanh'(x) = 1 - \tanh^2(x)),
ол да көп жағдайда 1-ден кіші. Сондықтан уақыт қадамдары көбейген сайын градиент әлсіреп, ертерек қадамдарға пайдалы сигнал жетпей қалады.

**2.4.** BPTT — RNN-ді уақыт бойымен “жазып ашып”, қателікті әр уақыт қадамы арқылы кері тарату әдісі. Feedforward желіге қарағанда RNN-ді оқыту қиынырақ, себебі бірдей параметрлер көп қадамда қайта қолданылады, ұзын тәуелділіктер бар, vanishing/exploding gradient жиі туады және есептеу толық параллельденбейді.

**3.1.** ([2, 4, 6, 8, 10]) қатарында модель ретті заңдылықты көреді: әр келесі мән алдыңғысынан 2-ге үлкен. RNN бұл қатарды қадамдап оқып, hidden state ішінде өсім үлгісін жинайды; соңғы контекст негізінде келесі мәнді 12-ге жақын деп болжайды. Мұнда негізгі идея — жеке санды емес, олардың ретпен өзгеру заңдылығын үйрену.

**3.2.** Мәтіндік деректерде input әдетте тізбек түрінде беріледі: символдар, сөздер не токендер алдымен индекске айналады, содан кейін модельге ретімен енгізіледі. One-hot encoding әр символды не сөзді сөздік көлеміндей вектормен белгілейді: бір позициясы 1, қалғаны 0. Бұл категориялық бірліктерді сандық форматқа көшіруге көмектеседі.

**3.3.**

1. Ұзын сөйлемдегі келесі сөзді болжау. Бастапқы бөліктегі маңызды сөз соңғы болжамға әсер етуі мүмкін, сондықтан ұзақ контекст керек.
2. Ауа температурасы, валюта бағамы сияқты ұзақ уақыттық қатарларды болжау. Бірнеше күн не апта бұрынғы үрдіс қазіргі мәнге ықпал етуі мүмкін.
3. Сөйлеуді тану. Дыбыс ағынында ертерек шыққан акустикалық белгі кейінгі интерпретацияға әсер етеді, сондықтан қысқа емес контекстті сақтау керек.
