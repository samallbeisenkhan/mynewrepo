№3 СӨЖ
Студенттердің аты мен бағалары бар сөздік жасап, ең жоғары бағаны тап


```python
students = {}

def add_student(name, grade):
    students[name] = grade
    print(f"Student {name} with grade {grade} added.")

def list_students():
    if not students:
        print("No students added yet.")
        return
    print("List of Students:")
    for name, grade in students.items():
        print(f"  {name}: {grade}")

def average_grade():
    if not students:
        return 0
    return sum(students.values()) / len(students)

def find_highest_grade():
    if not students:
        return None, None
    highest_grade = 0
    student_with_highest_grade = None
    for name, grade in students.items():
        if grade > highest_grade:
            highest_grade = grade
            student_with_highest_grade = name
    return student_with_highest_grade, highest_grade

if __name__ == '__main__':
    add_student('Aigerim', 85)
    add_student('Bulat', 92)
    add_student('Ali', 78)
    add_student('Dana', 95)
    list_students()
    print('Average:', average_grade())
    highest_student, highest_g = find_highest_grade()
    if highest_student:
        print(f'Student with the highest grade: {highest_student} ({highest_g})')
    else:
        print('No students to find the highest grade.')
```
