Тапсырма 2: Шартты Логика (Гранттың Базалық Мөлшерін Анықтау)
Шартты операторларды (if, elif, else) қолдана отырып, студенттің GPA-ға байланысты базалық грант мөлшерін анықтаңыз:
GPA ≥ 3.5 болса: 50000 (жоғары грант).
3.0 ≤ GPA < 3.5 болса: 25000 (орташа грант).
GPA < 3.0 болса: 0 (грант жоқ).
Базалық грант мөлшерін сақтайтын жаңа айнымалы құрыңыз.


```python
GPA = 3.85
if GPA >= 3.5:
  print('Сіздің бағаңыз: жоғары грант')
elif GPA >= 3.0 and GPA < 3.5:
  print('Сіздің бағаңыз: орташа грант')
else:
  print('Сіздің бағаңыз: грант жоқ')
```

```python
base_grant=0
def grant(gpa):
  if gpa>=3.5:
    base_grant=50000
  elif GPA>=3 and gpa<3.5:
    base_grant=25000
  else:
    base_grant=0
  return base_grant
grant(GPA)
```

Әлеуметтік жеңілдіктерді есептейтін функция құрыңыз.
Функцияның аты: social_bonus.
Функция бір аргумент қабылдауы керек: special_status (bool типі).
Егер special_status True болса, функция 10,000$ теңге қайтаруы керек, әйтпесе 0$ қайтаруы керек.
Функцияны шақырыңыз және нәтижесін жалпы грантқа қосыңыз.


```python
special_status = True
def social_bonus(special_status):
  if special_status == True:
    return 10000
  else:
    return 0
bonuses = social_bonus(special_status)
print(bonuses)
```
