Тапсырма 1: CSV Файлын Оқу және Тізімге Сақтау
Жүктелген Salaries.csv файлын with open(...) операторын қолданып оқыңыз.
Файлдың тек алғашқы 10 деректер жолын (data_rows атты тізімге) сақтаңыз. Тақырып жолын (header) санамаңыз.
data_rows тізімінің ұзындығын (len()) және бірінші элементін (бірінші қызметкердің дерегі) экранға шығарыңыз.


```python
import csv
data_rows = []
```

```python
f = open("Salaries (1).csv", "r")
print(f)
```

```python
with open('Salaries (1).csv', 'r', encoding='utf-8') as file:
  reader_object = csv.reader(file)
  header = next(reader_object)
  row_count = 0
  for row in reader_object:
            if row_count < 10:
                data_rows.append(row)
                row_count += 1
            else:
                break
  print("data_rows ұзындығы:", len(data_rows))
  print("Бірінші қызметкердің дерегі:", data_rows[0])
```

Тапсырма 2: Кортеж және Сөздік – Алғашқы 3 Адамның Атын Сақтау
data_rows тізімінен алғашқы 3 қызметкердің Атын (EmployeeName) алып, оларды Кортежге (first_three_names) сақтаңыз.
Осы 3 қызметкердің әрқайсысы үшін аты-жөнін (EmployeeName) Кілт ретінде, ал Жұмыс Атағын (JobTitle) Мән ретінде пайдаланып, Сөздік (employee_titles) құрыңыз.
Кортежді және Сөздікті экранға шығарыңыз.  


```python
first_three_names=(
    data_rows[0][1], data_rows[1][1], data_rows[2][1]
)
print(first_three_names)
```

```python
employee_titles = {}
for i in range(3):
  EmployeeName = data_rows[i][1]
  JobTitle = data_rows[i][2]
  employee_titles[EmployeeName] = JobTitle
print(employee_titles)
```

Тапсырма 3: Тізімдік Өрнектер (List Comprehension)
data_rows тізіміндегі 10 қызметкердің барлығының Базалық Жалақысын (BasePay) мәтіндік (str) түрде алып, оларды тізімдік өрнек (List Comprehension) арқылы base_pays_str атты жаңа тізімге жинаңыз.
Ескерту: Бұл жерде түрлендіру (float-қа) немесе сүзгілеу (filtering) қажет емес, тек мәтіндік деректі алу жеткілікті.
Жасалған base_pays_str тізімін экранға шығарыңыз.


```python
base_pays_str = [row[3] for row in data_rows]
base_pays_str
```

Тапсырма 4: Жиын (Set) – Бірегей Атақтарды Табу
data_rows тізіміндегі 10 қызметкердің барлығының Жұмыс Атақтарын (JobTitle) тізімдік өрнексіз, қарапайым for циклін пайдаланып алыңыз.
Бұл тізімді Жиын (unique_titles) құрылымына түрлендіріңіз.
Жиынның барлық элементтерін экранға шығарыңыз. Нелікwтен жиын пайдалы екенін қысқаша түсіндіріңіз (2 сөйлем).


```python
unique_titles=[]
for row in data_rows:
  unique_titles.append(row[2])
unique_titles

```

```python
set(unique_titles)
```

Set пайдалы, себебі ол элементтердің қайталануын автоматты түрде жояды және бірегей мәндермен жұмыс істеуге мүмкіндік береді. Сонымен қатар, set құрылымында іздеу, қосу және жою операциялары орташа есеппен өте жылдам орындалады.
